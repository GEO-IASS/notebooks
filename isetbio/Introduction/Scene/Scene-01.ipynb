## ISETBIO:  Scene

ISETBIO is structured around a few key objects that are important parts of the image acquisition pipeline. These are the scene, optical image, cone mosaic, bipolars, retinal ganglion cells.

A main goal of this script is to illustrate the ISETBIO programming style. By using ISETBIO objects properly, the code and analysis are much easier to understand. 

For each ISETBIO object type, there are three fundamental operations:  Create, set parameters, and get parameters.

Some objects were written before Matlab implemented its classes and thus doesn't rely on their organization. The more recent objects were written using classes.  The principles in both cases are the same.  Over time, we will convert everything to classes.

This script introduces the first and simplest ISETBIO object: a scene. The scene describes the radiance field.  For the present tutorial we  work with a simple planar radiance image, such as the image on a display surface.  


```matlab
% Initialize ISETBIOS
ieInit;
```

```matlab
% Create a simple scene of color patches
scene = sceneCreate('macbeth d65');
sceneShowImage(scene);
```

```matlab
% There are many different types of simple scenes available
scene = sceneCreate('rings rays');
sceneShowImage(scene);
```

```matlab
% You can also create a theoretical display and read in an RGB image
d = displayCreate('LCD-Apple');
fname = fullfile(isetRootPath,'data','images','rgb','eagle.jpg');
scene = sceneFromFile(fname,'rgb',[],d);
sceneShowImage(scene);
```
