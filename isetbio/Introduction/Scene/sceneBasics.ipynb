#### Introduction - Scene

An introduction to ISETBIO objects and functions: Scene

This script illustrates the ISETBIO programming style and describes the scene object.

ISETBIO is structured around a few key objects that are important parts of
the analysis. One of these is the scene.  Other key objects are
* image formation process (optical image)
* cone absorptions
*  cone photocurrent
* bipolar current
* ganglion cell spikes

By using ISETBIO objects properly, the code and analysis are much easier to understand. 

For each object there are three fundamental operations:  Create, set parameters, and get parameters.

The scene describes the spectral radiance field. For the present tutorial we  work with a simple planar radiance image, such as the image on a display surface.  

**Note**:
The implementation of some of these objects was written before Matlab implemented its classes.  Others were written using classes.  We will be converting everything to classes over time. But for now, this shouldn't matter to you.  The
principles are the same.


```matlab
% The create function initiates the object. 
%
% Typically, there are many different types of objects you can create. 
%
% sceneCreate builds a number of predefined scenes
%
% You can see the range of possibilities by typing
%
%    help sceneCreate
%
% into the Matlab prompt.

ieInit

% Create a scene: Here is a simple scene of a color chart
scene = sceneCreate('macbeth d65');

imagescRGB(sceneShowImage(scene));
```

```matlab
% Here is another

scene = sceneCreate('rings rays');

imagescRGB(sceneShowImage(scene));
```

```matlab
% Here is another

scene = sceneCreate('reflectance chart');

imagescRGB(sceneShowImage(scene));
```

```matlab
fname = fullfile(isetRootPath,'data','images','multispectral','StuffedAnimals_tungsten-hdrs.mat');
scene = sceneFromFile(fname,'multispectral');
imagescRGB(sceneShowImage(scene));
```
